## string methods

```python
text="HELLO WORLD"
```

```python
#len()-->returns the length of the string
text="HELLO WORLD"
print("the length of text:",len(text))
```

```python
print("lower case:",text.lower())
print("lower case:",text.upper())
```

```python
#strip()--> removes leading and trailing spaces)
text="      hello,world!"
print(text)
print(text.strip())
```

```python
#replace the word or character with anathor
text="HELLO WORLD"
print(text)
print("replace world with python",text.replace("world","python"))
print("---------------------------------------")
print(text.replace("L","a"))
```

```python
#split()----->split the string into list based on a seperator
text='apple,banana,mango'
print(text.split(","))
```

```python
# startseith()---> checks if string starts with a given prefix
text="Hello world"
print(text.startswith("He"))
```

```python
#endswith()--->checks string ends with a given suffix
print(text.endswith("d"))
```

```python
# find---> return the index of the first occurance of aa substring
print(text.find('o'))
```

```python
#count---> counts how many times a substrint occur
text="hello,hello,world"
print(text.count("hello"))
```

```python
# isalpha()---> checks if all characters are letters
text="Hello"
print(text.isalpha())
text1="Hello1"
print(text1.isalpha())
```

```python
# isdigit()---> checks if all characters are digits
te="123"
print(te.isdigit())
```

```python
# isalnum()----> checks if all charaters are letters or numbers
text="hello123"
print(text.isalnum())
```

```python
#capitalize---> it will give first character in uppercase
str.capitalize("hello world")
```

## format function

```python
name="prasad"
age="10"
print("my name is {} and i am {} years old".format(name,age))


```

```python
s="my name is {a} and i am {b} years old".format(a=name,b=age)

print(s)
```

```python
# %d ---->takes all the digit without decimals
# %f ---->takes all the digit with decimals
# %s  ---->takes string data


Integer=10
Float=10.6
String="prasad"
print("hi i am integer my value is %d\ni am float and my value is %f\ni am string and my value is %s"%(Integer,Float,String))
```

```python
# f-string -----> formated string
# place f before the string and use curly braces {} to insert variables /expressions

name ="john"
age=20
print(f"my name is {name} and i am {age} years old")
```

```python
a=2
b=3
print(f"the sum is {a+b}")
```

## list data type

```python
a=[10,'hello',3.14,1+3j]
print(a)
print(type(a))
```

```python
#indexing
a=[1,1,2,3,4,5]
print(a)
print(a[0])
print(a[1])
print(a[2])
print(a[3])
print(a[4])
```

```python
a=[1,1,2,3,4,5]
print(a[-1])
print(a[-2])
print(a[-3])
print(a[-4])

```

```python
#immutable---> we cannot change the element after forming
# sring is immutable
a="hello"
print(a[1])
print(a.replace('e','a'))
```

```python
#mutable---> we can change the element 
li=[1,2,3,]
li[1]="hello"
print(li)
```

```python
a=[1,1,2,3,4,5]
print(a[0:4])
print(a[::-1])
print(a[0:5:2])
print(a[-5:])
```

```python
a=[1,1,2,3,4,5]
a[0]="hello"
print(a)
```

```python
#nested list----> list inside the list
a=[10,20,30,[12,13]]
print(a)
print(a[3])
print(a[3][1])
```
